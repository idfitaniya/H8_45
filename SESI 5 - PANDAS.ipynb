# Conditional Formating

```python
df[["win_equiv"]].describe()
```

```python
df[df["win_equiv"]>42]
```

```python
df[(df["win_equiv"] > 42) & (df.year_id>2000)].describe()
```

```python
df[~(df["win_equiv"] > 42) & ~(df.year_id>2000)].describe()
```

```python
df[(df.lg_id=="ABA")]
```

```python
#TILDE SELAIN
df[~(df.lg_id=="ABA")]
```

```python
#ATAU |
df[(df["win_equiv"] > 42) | (df.year_id>2000)].describe()
```
